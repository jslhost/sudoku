```python
def create_list_of_squares(grid):
    list_of_squares = []
    start_row, end_row = 0, 3

    for i in range(3):
        start_col, end_col = 0, 3

        for j in range(3):
            list_of_squares.append(grid[start_row:end_row, start_col:end_col])
            start_col += 3
            end_col += 3
            
        start_row += 3
        end_row += 3

    return list_of_squares
```

```python
list_of_squares = create_list_of_squares(original_grid)
```

```python
def return_correct_square(i, j):
    if i < 3 and j < 3:
        return 0
    elif i < 3 and j < 6:
        return 1
    elif i < 3 and j < 9:
        return 2

    elif i < 6 and j < 3:
        return 3
    elif i < 6 and j < 6:
        return 4
    elif i < 6 and j < 9:
        return 5
    
    elif i < 9 and j < 3:
        return 6
    elif i < 9 and j < 6:
        return 7
    elif i < 9 and j < 9:
        return 8
```

```python
def find_adjacent_rows_and_cols(grid, i, j):
    if i in [0, 3, 6]:
        row_1, row_2 = (grid[i + 1], grid[i + 2])
    elif i in [1, 4, 7]:
        row_1, row_2 = (grid[i - 1], grid[i + 1])
    elif i in [2, 5, 8]:
        row_1, row_2 = (grid[i - 1], grid[i -2])

    if j in [0, 3, 6]:
        col_1, col_2 = (grid.T[j + 1], grid.T[j + 2])
    elif j in [1, 4, 7]:
        col_1, col_2 = (grid.T[j - 1], grid.T[j + 1])
    elif j in [2, 5, 8]:
        col_1, col_2 = (grid.T[j - 1], grid.T[j - 2])

    #---------------

    if j in [0, 1, 2]:
        in_1 = row_1[:3]
        out_1 = row_1[3:]
        in_2 = row_2[:3]
        out_2 = row_2[3:]
    elif j in [3, 4, 5]:
        in_1 = row_1[3:6]
        out_1 = np.concatenate((row_1[:3], row_1[6:]))
        in_2 = row_2[3:6]
        out_2 = np.concatenate((row_2[:3], row_2[6:]))
    elif j in [6, 7, 8]:
        in_1 = row_1[6:]
        out_1 = row_1[:6]
        in_2 = row_2[6:]
        out_2 = row_2[:6]


    if i in [0, 1, 2]:
        in_3 = col_1[:3]
        out_3 = col_1[3:]
        in_4 = col_2[:3]
        out_4 = col_2[3:]
    elif i in [3, 4, 5]:
        in_3 = col_1[3:6]
        out_3 = np.concatenate((col_1[:3], col_1[6:]))
        in_4 = col_2[3:6]
        out_4 = np.concatenate((col_2[:3], col_2[6:]))
    elif i in [6, 7, 8]:
        in_3 = col_1[6:]
        out_3 = col_1[:6]
        in_4 = col_2[6:]
        out_4 = col_2[:6]
    
    return (in_1, in_2, in_3, in_4), (out_1, out_2, out_3, out_4)
```
